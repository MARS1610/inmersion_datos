```python
inmuebles['Area'] = inmuebles['Area'].astype(float)
inmuebles['Precio'] = inmuebles['Precio'].astype(float)
precio = inmuebles['Precio'] / inmuebles['Area']
barrios = 10
p_barrio = precio.groupby(inmuebles['Barrio']).mean().sort_values(ascending=False).head(barrios)
```

```python
from seaborn.widgets import diverging_palette


plt.figure(figsize=(10,6))
plt.bar(p_barrio.index, p_barrio)
plt.xlabel('BARRIOS', fontsize=14, color='red')
plt.ylabel('PRECIO (pesos colombianos)', fontsize=14, color ='red')
plt.title('PRECIO DEL METRO CUADRADO POR BARRIO', fontsize=14,color='black')
plt.show()
```
